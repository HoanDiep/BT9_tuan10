```python
from keras.models import load_model
from google.colab import drive
drive.mount('/content/drive')
model100 = load_model('/content/drive/My Drive/Colab Notebooks/model_cifar100.h5')
```

```python
import matplotlib.pyplot as plt
from keras.utils import load_img,img_to_array
```

```python
url = '/content/drive/My Drive/Colab Notebooks/719144110651967980_000001.jpg'
img = load_img(url,target_size=(32,32))

plt.imshow(img)
```

```python
import numpy as np
img = img_to_array(img)
img = img.reshape(1,32*32*3)
img = img/255
np.argmax(model100.predict(img),axis=1)
```

```python

predicted_label = labels[model100.predict(img).argmax()]
print("Predicted label is {}".format(predicted_label))
```
